# Doubly Linked Lists

A Doubly Linked List is also known as a bi-directional list. It is comprised of `nodes` that keep and maintain a reference to the `previous` node in the collection, as well as the `next` node in the collection.

```python
# From scratch implementation of Doubly Linked Lists

class DoublyLinkedList:
    class __Node:
        def __init__(self, data):
            self.prev = None
            self.next = None
            self.data = data

    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, value):
        new_node = self.__Node(value)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node



    

    #INSERT
    # create a new node with provided value
    # if index is 0 then insert the new node at the beginning 
    #      link the new node to be the head
    #      if the new node is  the head
    #            the previous  head previous pointer is updated to the new node
    #      the new node is now set as the head of the list
    #      if not the tail, check if the tail is empty
    #            the new node now becomes the tail
    #      otherwise: 
    #            set the current to head
    #            set the count to 0
    #            start a loop that runs while the current and count are less than target index
    #                 previous is updated to point at current node
    #                 current pointer is set to next node to traverse
    #                 increment count by 1
    #            check to see if target index is found, node will insert here if true
    #                 new_node next pointer is set to current
    #                 new_node previous is set to previous
    #                 previous node next pointer is set to new_node
    #                 update next nodes previous pointer, if current: check if thhere is a node after new_node
    #                      if there is a node after new_node its previous pointer is updated to point back to new_node
    #                 otherwise: if current is none it means new node is being inserted at end of list
    #                 new node is set as tail of list
    #            otherwise 
    #                 append list if index is greater than the length of the list

        #this inserts a new node with the given value before the target index
        # Note: There are some additional considerations for the combinations of "index" and the state of our list
        def insert(self, index, value):
        new_node = self.__Node(value)
        if index == 0:
            new_node.next = self.head
            if self.head:
                self.head.prev = new_node
            self.head = new_node
            if not self.tail:
                self.tail = new_node
        else:
            current = self.head
            count = 0
            while current and count < index:
                previous = current
                current = current.next
                count += 1
            if count == index:
                new_node.next = current
                new_node.prev = previous
                previous.next = new_node
                if current:
                    current.prev = new_node
                else:
                    self.tail = new_node
            else:
                self.append(value)



    #REMOVE
    # check if list is empty by verifying that the head is none
        # return none if theres nothing to remove
    # traverse list by leaving current set to none
        # check if current node contains target value
            # update previous node pointer by checking if current node is not the head
                # if current node is not head, previous node next pointer  is set to point at node after current node.  (this removes from list)
                # otherwise if current node is the head, update the ead to point to next node
                # if current node is not the tail, the next node previous pinter is set as  current previous
                # otherwise if current node is tail set current previous to tail
                # after node is succesfully removed, return true
            # if current node does not contain valie, the current pointer is moved to next node and loop continues
            # if the loop completes without finding the value, return false
    
    #this removes the *first* instance of the given value
    def remove(self, value):
        if not self.head:
            return None

        current = self.head
        while current:
            if current.data == value:
                if current.prev:
                    current.prev.next = current.next
                else:
                    self.head = current.next

                if current.next:
                    current.next.prev = current.prev
                else:
                    self.tail = current.prev

                return True
            current = current.next
        
        return False




    #INDEX
    # start with the current set to head
    #      set the index to 0
    # traverse  the list
    #          if current data matches value
    #          return index
    #      otherwise
    #           move to the next node and +1 to index
    # if value is not found
    #      return error


        # this returns the *first* index of the given value
    def index(self, value):
        current = self.head
        count = 0
        while current:
            if current.data == value:
                return count
            current = current.next
            count += 1
        raise ValueError("Value not found in the list")




    
    #GET
    # check if the list is empty
        # if the list is empty return a index error
    # initialize traverse by setting current to point at ead
    # set count to 0
    # traverse the list with a loop while current is not None and count is less than target index
        # move the current pointer to the next node in the list to advance traverse
        # increment count by 1 as you move throughh  the list
    # return value if node is not found
        # check if current is not None which means the target index was found
            # return data stored in  current node
        # If not
            # raise an index error
    

    # this method allows you to use square bracket notation to retreive a value at the target index (IE dll[5])
    def __get__(self, index):
        if not self.head:
            raise IndexError("Index Error")
        
        current = self.head
        count = 0
        while current and count < index:
            current = current.next
            count += 1
        
        if current:
            return current.data
        else:
            raise IndexError("Index Error")


    #SET
    # check if the list is empty. 
        # if the list is empty it will raise an error
    # initialize a varible called current which starts by pointing to head
    # set a count to 0 to keep track of the current index as you traverse list
    # start a loop to traverse list
        # while current and count are less than  the index
        # move current pointer to next node, current = current.next
        # increment count by one, count += 1
    # set value at target index
        # if node is found and current data is equal to value
        # otherwise raise an index error
    
    def __set__(self, index, value):
            if not self.head:
                raise IndexError("Index Error")
            current = self.head
            count = 0
            while current and count < index:
                current = current.next
                count += 1
            if current:
                current.data = value
            else:
                raise IndexError("Index Error")



    # LENGTH
    # set count to 0
    # set current  node to head
    # start a loop while current is none
    # add an increment to keep track of nodes
    # after counting current node, current is updated to next to traverse list
    # once loop hhas finished, return the count

    
    def __len__(self):
        # this will return the number of Nodes in our list
        count = 0
        current = self.head
        while current:  
            count += 1
            current = current.next
        return count


    # BOOL
    # check if self is the  h ead
    # if self.head is not none, method returns True
    # if self.hhead is none method returns false

    def __bool__(self):
         # this will help evaluate our with with "if" and "if not"
        return self.head != None



    # INVERTED
    # create an empty list named values to store data
    # initialize a varible named current to point to tail of list - starting point for traversing list in reverse
    # start a loop while current is none, will travel from tail to head
    # add data to values list using repr(current.data) to ensure string representation of data is correctly formatted
    # current pointer is updated to previous node, continue reverse traverse
    # the list of values is combined into a string and seperated by comma and space.

    
    def inverted(self):
        #  this method allows us to print (or return  a string) of our list, but in inverted order
    values = []
    current = self.tail
    while current:
        values.append(repr(current.data))
        current = current.prev
    return "[" + ", ".join(values) + "]"



        
    def __str__(self):
        # remember, this method must return a string
        out = "["
        current = self.head
        if current:
            out += "%s" % repr(current.data)
            while current.next:
                current = current.next
                out += "], %s" % repr(current.data)
        out  += "]"
        return out
```

```python
# Comparing python3 lists to our DLL implementation 
pylist  = []
dll = DoublyLinkedList()

# Empty List :
print("Empty Python List: %s" % pylist)
print("Empty DLL: %s" % dll)

# Our lists with exactly one element
pylist.append(0)
dll.append(0)
print("Python list with one element: %s" % pylist)
print("DLL with one element: %s" % dll)

# Our list with more than one element
pylist.append(1)
dll.append(1)
print("Python list with more than one element: %s" % pylist)
print("DLL with more than one element: %s" % dll)

# Displaying the length of both lists
print("The length of our python list: %s" % len(pylist))
print("The length of our DLL: %s" % len(dll))
```
